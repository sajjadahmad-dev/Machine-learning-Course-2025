```python
import numpy as np 
import pandas as pd
from sklearn import linear_model
import math

def predict_using_sklean():
    df = pd.read_csv("test_scores.csv")
    model = linear_model.LinearRegression()
    model.fit(df[['math']], df.cs)
    return model.coef_, model.intercept_

def gradient_descent(x, y):
    m_start = 0
    b_start = 0
    iterations = 50
    n = len(x)
    learning_rate = 0.01
    cost_previous = 0

    for i in range(iterations):
        yp = m_start * x + b_start
        cost = (1/n) * sum([val**2 for val in (y - yp)])

        md = -(2/n) * sum(x * (y - yp))
        bd = -(2/n) * sum(y - yp)

        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break

        cost_previous = cost
        m_start = m_start - learning_rate * md
        b_start = b_start - learning_rate * bd

        print("m {}, b {}, cost {}, iteration {}".format(m_start, b_start, cost, i))

    return m_start, b_start

if __name__ == "__main__":
    df = pd.read_csv("test_scores.csv")
    x = np.array(df.math)
    y = np.array(df.cs)

    m, b = gradient_descent(x, y)
    print("Using gradient descent: Coef {} Intercept {}".format(m, b))

    m_sklearn, b_sklearn = predict_using_sklean()
    print("Using sklearn: Coef {} Intercept {}".format(m_sklearn, b_sklearn))
```

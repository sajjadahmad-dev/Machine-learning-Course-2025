Use iris flower dataset from sklearn library and use cross_val_score against following models to measure the performance of each. In the end figure out the model with best performance,


Logistic Regression

SVM

Decision Tree

Random Forest

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
X=iris.data
y=iris.target
```

```python
import numpy as np
```

# Logistic Regression

```python
log_reg = LogisticRegression(max_iter=200)

scores = cross_val_score(log_reg, X, y, cv=5)
print("Logistic Regression Accuracy:", scores)
print("Average Accuracy:", np.mean(scores))
```

# SVM

```python
svm = SVC()

scores_svm = cross_val_score(svm, X, y, cv=5)
print("SVM Accuracy:", scores_svm)
print("Average Accuracy Score for Logistic Regresion:", np.mean(scores_svm))
```

# Decision Tree

```python
dt = DecisionTreeClassifier()

scores_dt = cross_val_score(dt, X, y, cv=5)
print("Decision Tree Accuracy:", scores_dt)
print("Average Accuracy:", np.mean(scores_dt))
```

# Random Forest

```python
rf = RandomForestClassifier()

scores_rf = cross_val_score(rf, X, y, cv=5)
print("Random Forest Accuracy:", scores_rf)
print("Average Accuracy:", np.mean(scores_rf))
```

# In a loop

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "SVM": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5)
    print(name, "Average Accuracy:", np.mean(scores))
```

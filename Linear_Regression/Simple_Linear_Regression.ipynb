```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
plt.xlabel("year")
plt.ylabel("per capita income")
plt.scatter(df["year"],df["per capita income"],color="red",marker="+")
```

```python
X=df.drop("per capita income",axis=1)
```

```python
X
```

```python
y=df["per capita income"]
y
```

```python
# Create linear regression object
model = linear_model.LinearRegression()

# Train model
model.fit(X, y)

# Predict
prediction = model.predict([[2020]])

print(prediction)
```

```python
prediction=model.predict([[2020]])
```

```python
print(model.coef_)    # slope (m)
print(model.intercept_) # intercept (b)
```
